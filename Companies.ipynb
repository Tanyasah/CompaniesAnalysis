```python
p1['Number of companies']=p1['lead_name']
```

```python
p1=p1[['company_continent','Number of companies']]
p1
```

```python
p2=p2[['lead_continent','Number of Leads']]
p2
```

```python
fig=plt.figure()
ax0=fig.add_subplot(2,2,1)
ax1=fig.add_subplot(2,2,2)
ax2=fig.add_subplot(2,2,3)
ax3=fig.add_subplot(2,2,4)

list=['teal','steelblue','palegreen','skyblue','seagreen','darkgreen','lightseagreen']
list2=[0.1,0,0,0,0,0,0]
p1.plot('company_continent','Number of companies',kind='pie',autopct='%1.1f%%',figsize=(20,20),explode=list2,colors=list,labels=None,pctdistance=1.12,ax=ax0)
ax0.legend(labels=p1['company_continent'],loc='best')
ax0.set_title('PieChart of number of companies in various continents')

p1.plot('company_continent','Number of companies',kind='bar',ax=ax1,figsize=(20,25))
ax1.set_title("Bar chart of Number of Companies in various continents")
ax1.set_xlabel("Continents")
ax1.set_ylabel("Number of companies")

p2.plot('lead_continent','Number of Leads',kind='pie',autopct='%1.1f%%',figsize=(20,20),colors=list,explode=list2,labels=None,pctdistance=1.12,ax=ax2)
ax2.legend(labels=p2['lead_continent'],loc='best')
ax2.set_title('PieChart of number of Leads in various continents')

p2.plot('lead_continent','Number of Leads',kind='bar',figsize=(20,25),ax=ax3)
ax3.set_title("Bar chart of Number of Leads in various continents ")
ax3.set_xlabel("Continents")
ax3.set_ylabel("Number of Leads")
plt.show()
```
